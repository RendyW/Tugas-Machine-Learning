#### Handling Categorical Data

```python
# new dataset with categorical data

df2 = pd.read_csv('women_clothing_ecommerce_sales.csv')
df2.head()
```

```python
df2['size'].unique()
```

```python
df2['size'] = df2['size'].astype('category')
```

```python
# Remove rows that contain missing values in df2
df2 = df2.dropna(axis=0)
df2['size'].unique()
```

```python
# Mapping Ordinal Features

size_mapping = {'S': 1,
                'M': 2,
                'L': 3,
                'XL': 4,
                '2XL': 5,
                '3XL': 6,
                '4XL': 7,
                'One Size': 8,
                'One SIze': 8}

df2['size'] = df2['size'].map(size_mapping)
df2.head()
```

```python
# transform the integer values back to the ordinal string representation with Reverse Mapping Dictionary: inv_size_mapping

inv_size_mapping = {v: k for k, v in size_mapping.items()}
df2['size'].map(inv_size_mapping)
```

#### Encoding

```python
# Encoding with Mapping

# Create a mapping dictionary to be able to convert color from strings to integers
color_mapping = {label: idx for idx, label in enumerate(np.unique(df2['color']))}
color_mapping
```

```python
# Convert color in the datasets from strings to integers

df2['color'] = df2['color'].map(color_mapping)
df2.head(10)
```

```python
# reverse the color mapping (from integer to string) back to it's default value

inv_color_mapping = {v: k for k, v in color_mapping.items()}
df2['color'] = df2['color'].map(inv_color_mapping)
df2.head(10)
```

```python
# Encoding with LabelEncoder

# import LabelEncoder
from sklearn.preprocessing import LabelEncoder

# Label encoding with sklearn's LabelEncoder
color_labelEncoder = LabelEncoder()
y = color_labelEncoder.fit_transform(df2['color'].values)
y

# The fit_transform method became a shortcut for calling fit and transform separately
```

```python
# Use inverse_transform method to transform integer back to the default value (string)

color_labelEncoder.inverse_transform(y)
```

#### One Hot Encoding

```python
X = df2[['color', 'size', 'unit_price']].values
color_le = LabelEncoder()
X[:, 0] = color_le.fit_transform(X[:, 0])
X

# notes: X[:, 0] means that u take every row and 0 column from X(the dataset).
```

```python
# Performing one hot encoding to the row of the datasets

# import OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

X = df2[['color', 'size', 'unit_price']].values
color_ohe = OneHotEncoder()
color_ohe.fit_transform(X[:, 0].reshape(-1, 1)).toarray()
```

```python
# Performing one hot encoding to the column of the datasets

# import ColumnTransformer
from sklearn.compose import ColumnTransformer

X = df2[['color', 'size', 'unit_price']].values
c_transf = ColumnTransformer([('onehot', OneHotEncoder(), [0]),
                              ('nothing', 'passthrough', [1,2])])
c_transf.fit_transform(X).astype(float)
```

```python
# one hot encoding via pandas

pd.get_dummies(df2[['unit_price', 'color', 'size']]).head()
```

```python
# multicollinearity guard in get_dummies

pd.get_dummies(df2[['unit_price', 'color', 'size']], drop_first=True).head()
```

```python
# multicollinearity guard for the OneHotEncoder
color_ohe = OneHotEncoder(categories='auto', drop='first')
c_transf = ColumnTransformer([('onehot', color_ohe, [0]),
                              ('nothing', 'passthrough', [1,2])])
c_transf.fit_transform(X).astype(float)
```

#### Partitioning Dataset in Training and Test Sets

```python
df = pd.read_csv('train.csv')
print('UserID', np.unique(df['User_ID']))
```

```python
df.head()
```

```python
# select the numeric data from df
df = df.select_dtypes([np.number])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X, y = df.iloc[:, 1:].values, df.iloc[:, 0].values

X_train, X_test, y_train, y_test =\
    train_test_split(X, y,
                     test_size=0.3,
                     random_state=0,
                     stratify=y)

# notes: df.iloc[:, 1:] -> take all rows and columns starting from index number 1
# notes: df.iloc[:, 0] -> take all rows and 1 column from index number 0
```

#### Bringing Features Onto the Same Scale

```python
# Normalization

from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
X_train_norm = mms.fit_transform(X_train)
X_test_norm = mms.transform(X_test)
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
```
